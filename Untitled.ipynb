```python
import datalabframework as dlf
```

```python
dlf.engine('spark')
```

```python
e = dlf.engine()
```

```python
e.load('data/raw/olist_customers_dataset.csv').data.grid(render='pandas')
```

```python
df = e.load('data/raw/olist_orders_dataset.csv').join(e.load('data/raw/olist_customers_dataset.csv'), on='customer_id', how='left')
```

```python
df = ( df
        .cols.get('customer_id').drop()
        .cols.get('customer_unique_id').rename('cid')
     )
```

```python
df.cols.order('cid', 'order_id', 'order_status').data.grid(render='pandas')
```

```python
df = e.load('data/raw/olist_orders_dataset.csv')
```

```python
df.cols.groupby('customer_id').agg('count').cols.find('count').agg('count', 'min', 'max', 'avg').data.grid(render='pandas')
```
